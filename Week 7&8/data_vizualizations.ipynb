For this assignment, I'm going to use a random dataset about Cars in USA which I found on Kaggle. US Cars'data was scraped from AUCTION EXPORT.com. This dataset included Information about 28 brands of clean and used vehicles for sale in US. Twelve features were assembled for each car in the dataset. <br>
Here's the link to the Kaggle dataset source: https://www.kaggle.com/doaaalsenani/usa-cers-dataset

```python
# necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the dataset ie in the same working directory
df = pd.read_csv("USA_cars_datasets.csv")
df.head()
```

```python
df.info()
```

Let's see the brands of car and its distribution

```python
df.describe()
```

```python
# Car brands and their count in a bar diagram
fig = plt.figure(figsize=(12, 5))
ax = fig.add_axes([0,0,1,1])
sns.countplot(x='brand', data=df, ax=ax)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")

plt.show()
```

```python
# scatter plot of mileage of cars and their price
plt.figure(figsize=(8,6))
sns.scatterplot(x='mileage',y='price',data=df,color='purple')
```

```python
# Scatter plot with year and price
plt.figure(figsize=(8,5))
sns.scatterplot(x='year',y='price',data=df,hue='title_status',alpha=0.2)
```

```python
# Mileage distribution of cars
sns.set_style('whitegrid')
plt.figure(figsize=(8,5))
sns.histplot(df['price'])
```
